```python
import random
random.seed(0)
org_data=list(range(-50,50))
random.shuffle(org_data)
print('org_data',org_data)
```

# 01 冒泡排序

```python
def bubbleSort(data):
    n=len(data)
    swap=False
    for i in range(n)[::-1]:
        for j in range(i):
            if data[j]>data[j+1]:
                swap=True
                data[j],data[j+1]=data[j+1],data[j]
        if not swap:break
    return data
```

```python
sort_data=bubbleSort(org_data.copy())
print('sort_data:',sort_data)
```

# 02 选择排序

排序过程：<br />
    1.首先通过n-1次关键字比较，从n个记录中找出关键字最小的记录，将它与第一个记录交换<br />
    2.再通过n-2次比较，从剩余的n-1个记录中找出关键字次小的记录，将它与第二个记录交换<br />
    3.重复上述操作，共进行n-1趟排序后，排序结束<p>
    
无论什么数据进去都是 O(n²) 的时间复杂度。所以用到它的时候，数据规模越小越好。唯一的好处可能就是不占用额外的内存空间了吧。

```python
def selectionSort(data):
    '''
    选择排序
    '''
    n=len(data)
    for i in range(n):
        for j in range(i+1,n):
            if data[j]<data[i]:
                data[j],data[i]=data[i],data[j]
    return data
```

```python
sort_data=selectionSort(org_data.copy())
print('sort_data:',sort_data)
```

# 03 插入排序

排序过程：<br />
    1.先将序列中第1个记录看成是一个有序子序列，然后从第2个记录开始，逐个进行插入，直至整个序列有序；<br />
    2.整个排序过程为n-1趟插入。<br />

```python
def insertSort(data):
    '''
    插入排序
    '''
    n=len(data)
    for i in range(1,n):
        temp=data[i]
        index=i-1
        while index>=0 and data[index]>temp:
            data[index+1]=data[index]
            index-=1
        data[index+1]=temp
    return data
```

```python
sort_data=insertSort(org_data.copy())
print('sort_data:',sort_data)
```

# 04 希尔排序

<img src="./5.png" width = "300" height = "200" alt="图片名称" align='left'/>

希尔排序可提高排序速度，因为<p>
1.直接插入排序T(n)=O(n²)，分组后n值减小，n²更小，而所以T(n)减小了<br />
2.关键字较小的记录跳跃式前移，在进行最后一趟增量为1的插入排序时，序列已基本有序<br />

```python
def shellSort(data):
    '''
    希尔排序
    '''
    gap=1
    n=len(data)
    while(gap < n/3):
        gap = gap*3+1
    while gap>0:

        for i in range(gap,n):
            temp=data[i]
            index=i-gap
            while index>=0 and data[index]>temp:
                data[index+gap]=data[index]
                index-=gap
            data[index+gap]=temp
                
        gap//=3
    return data
```

```python
sort_data=shellSort(org_data.copy())
print('sort_data:',sort_data)
```

# 05 归并排序

```python
def mergeSort(data):
    '''
    归并排序，迭代实现
    '''
    n=len(data)
    step=0
    
    while 2**step<=n:
        curr_data=[]
        left_end,right_end=2**step-1,2*2**step-1   #要合并两块的末尾
        while left_end<n-1:   #保证有左右两块可以合并
            
            left,right=left_end-2**step+1,right_end-2**step+1    #要合并两块的起始
            right_end=min(n-1,right_end)
            
            while left<=left_end and right<=right_end:   #合并相邻两块
                if data[left]<=data[right]:
                    curr_data.append(data[left])
                    left+=1
                else:
                    curr_data.append(data[right])
                    right+=1
            
            curr_data+=data[left:left_end+1]     #将两块中未被合并的剩余部分直接加到后面
            curr_data+=data[right:right_end+1]
            
            left_end+=2*2**step
            right_end+=2*2**step
            
        curr_data+=data[left_end-2**step+1:]   #将未被合并的剩余部分直接加到后面
        data=curr_data
        step+=1
        
    return data
```

```python
sort_data=mergeSort(org_data.copy())
print('sort_data:',sort_data)
```

```python
def mergeSort(data):
    '''
    归并排序，递归实现
    '''
    n=len(data)
    if n<=1:return data
    return merge(mergeSort(data[:n//2]),mergeSort(data[n//2:]))
def merge(left,right):
    res=[]
    left_len,right_len=len(left),len(right)
    left_idx,right_idx=0,0
    while left_idx<=left_len-1 and right_idx<=right_len-1:
        if left[left_idx]<=right[right_idx]:
            res.append(left[left_idx])
            left_idx+=1
        else:
            res.append(right[right_idx])
            right_idx+=1
            
    res+=left[left_idx:]     #将两块中未被合并的剩余部分直接加到后面
    res+=right[right_idx:]
    return res
```

```python
sort_data=mergeSort(org_data.copy())
print('sort_data:',sort_data)
```

# 06 快速排序

快速排序通常明显比其他 Ο(nlogn) 算法更快，因为它的内部循环（inner loop）可以在大部分的架构上很有效率地被实现出来。
且 O(nlogn) 记号中隐含的常数因子很小，比复杂度稳定等于 O(nlogn) 的归并排序要小很多。所以，对绝大多数顺序性较弱的随机数列而言，快速排序总是优于归并排序。

```python
def quickSort(data):
    '''
    快速排序
    '''
    n=len(data)
    if n<=1:return data
    pivot=data[0]    #中枢数值，默认取第一个点为中枢
    left,right=0,n-1
    right_go=True   #是否由右指针移动
    while left<right:  #左右指针指向同一位置时停止
        if right_go:
            if data[right]>=pivot:
                right-=1
            else:
                data[left]=data[right]
                left+=1
                right_go=False
        else:
            if data[left]>=pivot:
                data[right]=data[left]
                right-=1
                right_go=True
            else:
                left+=1
    return quickSort(data[:left])+[pivot]+quickSort(data[left+1:])
```

```python
sort_data=quickSort(org_data.copy())
print('sort_data:',sort_data)
```

# 07 堆排序

筛选方法：<br />
    输出堆顶元素(根)之后，以堆中最后一个元素替代之；<br />
    然后将根结点值与左、右子树的根结点值进行比较，并与其中小(大)者进行交换；<br />
    重复上述操作，直至叶子结点，将得到新的堆，称这个从堆顶至叶子的调整过程为“筛选”<p>
初建堆方法：<br />
    从无序序列的第n//2个元素（即此无序序列对应的完全二叉树的最后一个非叶子结点）起，至第一个元素止，进行反复筛选<br />

<img src="./6.png" width = "500" height = "300" alt="图片名称" align='left'/>

```python
def buildMinHeap(data,m):
    '''
    筛选，从第m个节点开始建立小顶堆（节点编号从1开始）
    '''        
    if len(data) >= 2*m+1 and min(data[2*m-1],data[2*m])<data[m-1]:
        if data[2*m-1]<=data[2*m]:
            data[m-1],data[2*m-1]=data[2*m-1],data[m-1]
            buildMinHeap(data,2*m)
        else:
            data[m-1],data[2*m]=data[2*m],data[m-1]
            buildMinHeap(data,2*m+1)
        
    elif len(data) == 2*m and data[2*m-1]<data[m-1]:
        data[m-1],data[2*m-1]=data[2*m-1],data[m-1]
        buildMinHeap(data,2*m)
        
    return data
def heapSort(data):
    '''
    堆排序
    '''
    n=len(data)
    if n==0:return data
    for i in range(1,n//2+1)[::-1]:   #从第n//2个元素开始到第1个元素，反复筛选
        data=buildMinHeap(data,i)
    res=[]

    while True:
        res.append(data.pop(0))
        if data==[]:break
        data.insert(0,data.pop())
        data=buildMinHeap(data,1)

    return res
```

```python
sort_data=heapSort(org_data.copy())
print('sort_data:',sort_data)
```

# 08 桶排序

为了使桶排序更加高效，我们需要做到这两点：<br />
    在额外空间充足的情况下，尽量增大桶的数量。<br />
    使用的映射函数能够将输入的 N 个数据均匀的分配到 K 个桶中。<p>


同时，对于桶中元素的排序，选择何种比较排序算法对于性能的影响至关重要。<p>

什么时候最快<br />
    当输入的数据可以均匀的分配到每一个桶中。<p>

什么时候最慢<br />
    当输入的数据被分配到了同一个桶中。<p>

```python

def bucketSort(data):
    """桶排序"""
    min_num = min(data)
    max_num = max(data)
    # 桶的大小
    bucket_range = (max_num-min_num) / len(data)
    # 桶数组
    count_list = [ [] for _ in range(len(data) + 1)]
    # 向桶数组填数
    for num in data:
        count_list[int((num-min_num)//bucket_range)].append(num)
    data.clear()
    # 回填，这里桶内部排序直接调用了sorted
    for bucket in count_list:
        data+=quickSort(bucket)   #这里对每个桶内数据进行快速排序
    return data
```

```python
sort_data=bucketSort(org_data.copy())
print('sort_data:',sort_data)
```

# 09 链式基数排序

数排序 vs 计数排序 vs 桶排序<p>

这三种排序算法都利用了桶的概念，但对桶的使用方法上有明显差异：<br />
    基数排序：根据键值的每位数字来分配桶；<br />
    计数排序：每个桶只存储单一键值；<br />
    桶排序：每个桶存储一定范围的数值。<br />

```python
def linkedRadixSort(data):
    '''
    链式基数排序
    '''
    i=0
    
    while True:
        bucket_plus={}
        bucket_minus={}
        for num in data:
            label=int((abs(num)/10**i)%10)
            if num>=0:
                bucket_plus[label]=bucket_plus.get(label,[])
                bucket_plus[label].append(num)
            else:
                bucket_minus[label]=bucket_minus.get(label,[])
                bucket_minus[label].append(num)
            
        data.clear()

        for label in range(10)[::-1]:
            data+=bucket_minus.get(label,[])
        for label in range(10):
            data+=bucket_plus.get(label,[])
        
        if len(bucket_minus)==1 and len(bucket_plus)==1:break

        i+=1
    return data
```

```python
sort_data=linkedRadixSort(org_data.copy())
print('sort_data:',sort_data)
```
